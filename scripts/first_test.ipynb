```python
import sys
sys.path.append('../utils')
from datagenerators import ODESimulator, SIR
import torch
import numpy as np
import matplotlib.pyplot as plt
```

### Generate data

```python
# Data
beta, gamma = 1/2, 1/3
X0 = torch.tensor([0.99, 0.01, 0.0]).reshape(-1, 3)

Nd = 101 # Number of data points

time_int = [0, 40]
spacing = (time_int[1] - time_int[0]) / Nd

data_time_points = np.linspace(time_int[0], time_int[1], Nd)

data = ODESimulator(SIR, (beta, gamma), X0, data_time_points)

td = data.td
Xd = data.Xd

data.Xd.shape
```

```python
plt.plot(data_time_points, Xd[:, 0], label='S')
plt.plot(data_time_points, Xd[:, 1], label='I')
plt.plot(data_time_points, Xd[:, 2], label='R')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.title('SIR Model Simulation')
plt.show()
```
